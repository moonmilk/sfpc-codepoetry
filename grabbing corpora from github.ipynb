# playing with Darius' corpora
For a quick experiment, load a wordlist straight from github

```python
# http://stackoverflow.com/a/12965254
import urllib, json
url = 'https://github.com/dariusk/corpora/raw/master/data/music/genres.json'
response = urllib.urlopen(url);
genres = json.loads(response.read())
print genres

```

```python
[g for g in genres['genres'] if 'punk' in g or 'Punk' in g]
```

```python
fluids = json.loads(urllib.urlopen('https://github.com/dariusk/corpora/raw/master/data/materials/abridged-body-fluids.json').read())
```

```python
fluids
```

```python
import re, random

for genre in genres['genres']:
    m = re.sub('(?i)(rock|pop|punk|metal|house|trance|doom|grunge|sound|core)', random.choice(fluids['abridged body fluids']), genre)
    print m
```
