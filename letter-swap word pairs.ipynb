# messing with words
Most unix systems (including Mac OSX) have a somewhat eccentric list of words in `/usr/share/dict/words` or `/usr/dict/words`

```python
# Much faster to check if a word is in a dict than in a list
with open('/usr/share/dict/words') as f:
    words = dict([(w.strip(), True) for w in f.readlines()])
```

```python
# check that the words dict is doing what you expect
'yes' in words, 'nopes' in words
```

```python
# for each word, swap every pair of letters and see if the result is also a word
pairs = []
for w in words:
    for i in range(0,len(w)):
        for j in range(i+1,len(w)):
            word = list(w)
            word[i],word[j] = word[j],word[i]
            wrod = "".join(word)
            if w != wrod and wrod in words:
                pairs.append([w,wrod])
            
```
